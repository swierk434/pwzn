```python
import matplotlib.pyplot as plt
```

```python
class SIR:
    def __init__(self, Istart = 0.01, Rstart = 0, beta = 0.7, gamma = 0.1, dt = 0.01, n = 10000):
        self.beta = beta
        self.gamma = gamma
        self.dt = dt
        self.I = Istart
        self.R = Rstart
        self.S = 1 - (Istart + Rstart)
        self.t = 0
        self.Stab = []
        self.Itab = []
        self.Rtab = []
        self.Ttab = []
        self.simulate(n)
        self.print_chart_plt()

    def single_timestep(self):
        dS = -self.beta*self.S*self.I*self.dt
        dR = self.gamma*self.I*self.dt
        self.S += dS
        self.I -= (dS+dR)
        self.R += dR
        self.t += self.dt

    def simulate(self, n):
        for m in range(n):
            #print(self.S)
            self.Stab.append(self.S)
            self.Itab.append(self.I)
            self.Rtab.append(self.R)
            self.Ttab.append(self.t)
            self.single_timestep()

    def print_chart_plt(self):
        fig = plt.figure(figsize=(10, 6))
        plt.plot(self.Ttab, self.Stab,label='S', color='green')
        plt.plot(self.Ttab, self.Itab, label='I', color='red')
        plt.plot(self.Ttab, self.Rtab, label='R', color='blue')
        plt.ylabel('% of population')
        plt.xlabel('t')
        plt.title(f'Model SIR beta = {self.beta}, gamma = {self.gamma}')
        plt.legend()
        plt.grid(True)
        plt.show()
```

```python
sir1 = SIR()
sir = SIR(Istart = 0.01, Rstart = 0, beta = 0.13, gamma = 0.1, dt = 0.1, n = 10000)
sir = SIR(Istart = 0.01, Rstart = 0, beta = 0.2, gamma = 0.1, dt = 0.02, n = 10000)
sir = SIR(Istart = 0.3, Rstart = 0, beta = 0.13, gamma = 0.1, dt = 0.1, n = 10000)
sir = SIR(Istart = 0.3, Rstart = 0, beta = 0.2, gamma = 0.1, dt = 0.02, n = 10000)
```
