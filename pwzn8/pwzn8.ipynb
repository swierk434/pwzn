```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
sp.init_printing()
```

```python
x = sp.Function('x')
F = sp.Function('F')
t = sp.Symbol('t')
w0 = sp.Symbol('omega_0', positive=True, real = True)
wr = sp.Symbol('omega_r', positive=True, real = True)
A = sp.Symbol('A', real = True)
fi = sp.Symbol('phi', real = True)
b = sp.Symbol('beta', positive = True, real = True)
x0 = sp.Symbol('x_0', real = True)
v0 = sp.Symbol('v_0', real = True)

F = A*sp.cos(wr*t+fi)
```

```python
eq2 = sp.Eq(x(t).diff(t,t), -w0**2 * x(t) - b*x(t).diff(t))
eq3 = sp.Eq(x(t).diff(t,t), -w0**2 * x(t) + F)    
```

```python
eq2
```

```python
eq3
```

```python
sol2 = sp.dsolve(eq2, x(t), ics={x(0): x0, x(t).diff(t).subs(t, 0): v0})
sol3 = sp.dsolve(eq3, x(t), ics={x(0): x0, x(t).diff(t).subs(t, 0): v0})
```

```python
sp.plot(sol2.rhs.subs({x0: 1, v0 : -1, w0 : 5, b : 5}), (t, 0 ,10))
```

```python
sp.plot(sol2.rhs.subs({x0: 1, v0 : -1, w0 : 5, b : 0.5}), (t, 0 ,10))
```

```python
sp.plot(sol3.rhs.subs({x0: 1, v0 : -1, w0 : 2, wr : 3, fi : 0, A : 1}), (t, 0, 40))
```
